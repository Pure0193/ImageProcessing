```python
# Load training images
import os
import cv2 as cv
from sklearn.svm import LinearSVC
from skimage import feature

images = []
labels = []
image_paths = os.listdir(f"/content/drive/MyDrive/Colab Notebooks/dataset/faces94")
# Try first 20 subjects
s = 0

for path in image_paths:
  s=s+1
  if (s > 20):
    break
  # Get all folders (i.e. subjects)
  all_images = os.listdir(f"/content/drive/MyDrive/Colab Notebooks/dataset/faces94/{path}")
  
  # Get all images and their labels, for each subject
  # First 10 images for training
  # Last 10 images for validating
  c = 0
  for image in all_images:
    c=c+1
    if (c > 10):
      break
    image_path = f"/content/drive/MyDrive/Colab Notebooks/dataset/faces94/{path}/{image}"
    image = cv.imread(image_path)
    image = cv.resize(image, (180, 200))
    # Get HOG descriptors for each image
    hog_desc = feature.hog(image, orientations=9, pixels_per_cell=(8,8),
                           cells_per_block=(2,2), transform_sqrt=True, block_norm='L2-Hys')
    # Update the training data and labels
    images.append(hog_desc)
    labels.append(path)
```

```python
# Train
svm_model = LinearSVC(C=10)
svm_model.fit(images, labels)
```

```python
# Predicting test images
# Try first 20 subjects
image_paths = os.listdir(f"/content/drive/MyDrive/Colab Notebooks/dataset/faces94")
s = 0
acc = 0
total = 0

dir = f"/content/drive/MyDrive/Colab Notebooks/dataset/faces94/results"
if not os.path.exists(dir):
  os.mkdir(dir)

for path in image_paths:
  s=s+1
  if (s > 20):
    break
  # Get all folders (i.e. subjects)
  all_images = os.listdir(f"/content/drive/MyDrive/Colab Notebooks/dataset/faces94/{path}")
  dir = f"/content/drive/MyDrive/Colab Notebooks/dataset/faces94/results/{path}"
  if not os.path.exists(dir):
    os.mkdir(dir)

  # Get all images and their labels, for each subject
  # First 10 images for training
  # Last 10 images for validating
  c = 0
  for image in all_images:
    c=c+1
    if (c <= 10):
      continue
    total+=1
    image_path = f"/content/drive/MyDrive/Colab Notebooks/dataset/faces94/{path}/{image}"
    image_in = cv.imread(image_path)
    resized_image = cv.resize(image_in, (180, 200))
    # Get HOG descriptor for each test image
    hog_desc = feature.hog(resized_image, orientations=9, pixels_per_cell=(8,8),
                           cells_per_block=(2,2), transform_sqrt=True, block_norm='L2-Hys')
    # Prediction
    pred = svm_model.predict(hog_desc.reshape(1,-1))[0]
    print(path, pred.title())
    if(path.lower() == pred.title().lower()):
      acc+=1
    # Put a predicted text on each test image
    cv.putText(image_in, pred.title(), (20,40), cv.FONT_HERSHEY_SIMPLEX, 1.0, (0, 255, 0), 2)
    cv.imwrite(f"/content/drive/MyDrive/Colab Notebooks/dataset/faces94/results/{path}/{image}", image_in)
acc = acc*100/total
print(acc)
```
